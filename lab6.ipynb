```python
"""
Authors: Viktoriia Vlasenko
"""
import numpy as np
import matplotlib.pyplot as plt


def F(x, y):
    return (y-x**2)

def Fanalityczna(x):
    return (2 + 2*x + x**2 - np.exp(x))


#krok
h = 1
fig = plt.figure()
ax = plt.axes()


x = np.arange(0, 3+h, h)
xa = np.arange(0, 3, 0.005)
y = np.zeros(len(x))
y[0] = 1
for i in range(len(x)-1):
    y[i + 1] = y[i] + h*F(x[i], y[i])
    
ax.plot(xa, Fanalityczna(xa), label='dane analityczne')
ax.scatter(x, y, label='Euler')

    
for i in range(len(x)-1):
    k1 = F(x[i], y[i])
    k2 = F(x[i] + 0.5*h, y[i] + 0.5*h*k1)
    k3 = F(x[i] + 0.5*h, y[i] + 0.5*h*k2)
    k4 = F(x[i] + h, y[i] + h*k3)
    y[i + 1] = y[i] + (k1 + 2*k2 + 2*k3 + k4)*h/6
    

ax.scatter(x, y, label='RK IV rzad')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.legend()
```
